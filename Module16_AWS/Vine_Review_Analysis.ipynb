```python
# Import Dependencies
import pandas as pd
```

```python
# import data
vine_df = pd.read_csv("vine_table.csv")
vine_df
```

```python
vine_df_20 = vine_df.loc[vine_df["total_votes"] >= 20]
vine_df_20
```

```python
vine_df_50 = vine_df_20.loc[(vine_df_20["helpful_votes"] / vine_df_20["total_votes"]) >= 0.50]
vine_df_50
```

```python
paid_reviews = vine_df.loc[vine_df["vine"] == "Y"]
paid_reviews
```

```python
unpaid_reviews = vine_df.loc[vine_df["vine"] == "N"]
unpaid_reviews
```

```python
paid_5star_reviews = paid_reviews.loc[paid_reviews["star_rating"] == 5]["star_rating"].count()
paid_5star_reviews
```

```python
paid_reviews_count = paid_reviews["star_rating"].count()
paid_reviews_count
```

```python
percent_paid_5star = round((paid_5star_reviews / paid_reviews_count) * 100, 2)
percent_paid_5star
```

```python
unpaid_reviews_count = unpaid_reviews["star_rating"].count()
unpaid_reviews_count
```

```python
unpaid_5star_reviews = unpaid_reviews.loc[unpaid_reviews["star_rating"] == 5] ["star_rating"].count()
unpaid_5star_reviews
```

```python
percent_unpaid_5star = round((unpaid_5star_reviews / unpaid_reviews_count) * 100, 2)
percent_unpaid_5star
```
